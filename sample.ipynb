```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import neighbors, datasets
iris = datasets.load_iris()
iris
```

```python
iris_X = iris.data
iris_Y = iris.target
print(iris_X)
print(iris_Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(iris_X, iris_Y, test_size=50)
print(len(y_train))
print(len(y_test))
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=1, p=2)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
clf = neighbors.KNeighborsClassifier(n_neighbors=10, p=2)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(y_pred)
print(accuracy_score(y_test, y_pred))
```

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('./Mall_Customers.csv')
df.head()
```

```python
x = df.iloc[:,[3,4]].values
x
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x_scale = scaler.transform(x)
x_scale
```

```python
from sklearn.cluster import KMeans
wcss_list = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x_scale)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(x_scale)
y_predict = kmeans.predict(x_scale)
y_predict
```

```python
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], c='red', s = 100, label='Cluster 1')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], c='blue', s = 100, label='Cluster 2')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], c='green', s = 100, label='Cluster 3')
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], c='yellow', s = 100, label='Cluster 4')
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') # Cụm 5

kmeans.cluster_centers_ = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='yellow', s = 200, label='Centroid')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
